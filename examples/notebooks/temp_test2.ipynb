```python
@interact(nodecolor=['lightblue','lightgreen','lightcoral'],
          nodelabelcolor=['yellow','white','black'],
          fontscale=(0.5,3.0,0.1),
          nodeoutlinecolor=['salmon','blue','green','black'],
          drawselfloop=[True,False],
          selfloopangle=(0,360,15),
          selflooplabelscale=(0.5,3.0,0.1),
          selfloopcolor=['darkred','darkgreen','darkblue'],
          )

def draw_selfloop_nodes(nodecolor='lightblue',
                        nodelabelcolor='yellow',
                        fontscale=1.5,
                        nodeoutlinecolor='salmon',
                        drawselfloop=True,
                        selfloopangle=90,
                        selflooplabelscale=1,
                        selfloopcolor='darkred',
                        ):
    graph = gp.GraphCircuit()

    x = np.arange(-30,30,8)

    for idx,xx in enumerate(x):
        graph.addnode(label=f'{idx}',xy=(xx,0),
                    nodecolor=nodecolor,
                    nodelabelcolor=nodelabelcolor,
                    drawselfloop=drawselfloop,
                    R = 1.5,
                    fontscale=fontscale,
                    nodelabelnudge=(-0.05,-0.1),
                    nodeoutlinecolor=nodeoutlinecolor,
                    selflooplabel=r'$\Delta_\mathrm{A}$',
                    selfloopangle=selfloopangle,
                    selflooplabelscale=selflooplabelscale,
                    selfloopcolor=selfloopcolor,
                    )
    graph.draw()
    plt.show()
    # plt.axis('on')
    # plt.grid(True)
```

```python
def ABgraph(
    conj = True,
    sep = 10,
    fontscale = 0.75,
    Alabelnudgex = 0,Alabelnudgey = 0,
    Blabelnudgex = 0,Blabelnudgey = 0,
    edgelabeloffset = 2.3,
    edgelw = 3,
    style = 'og',
    figsize = 11,
    ):  

    # global g

    if conj:
        betalabel = r'$-\beta_{BA}^*$'
        tonode = 'B*'
    else:
        betalabel = r'$\beta_{BA}^*$'
        tonode = 'B'

    if style == 'og':
        drawselfloop = True
        edgestyle = 'loopy'
    elif style == 'ofer':
        drawselfloop = True
        edgestyle = 'double'
        betalabel = None
    elif style == 'simple':
        drawselfloop = False
        edgestyle = 'double'
        betalabel = None


    Alabel = 'A'
    Blabel = 'B'

    g = gp.GraphCircuit()
    g.addprettynode(mode = Alabel,
                    xy = (0,0),
                    D = 1.1,
                    loopangle = 180,
                    drawselfloop = drawselfloop,
                    fontscale=fontscale,
                    nodelabelnudge = (Alabelnudgex,Alabelnudgey),
                    )

    g.addprettynode(mode = Blabel,
                    xy = (sep,0),
                    D = 1.1, 
                    conj=conj,
                    fontscale=0.8,
                    loopangle = 0,
                    drawselfloop = drawselfloop,
                    nodelabelnudge = (Blabelnudgex,Blabelnudgey),
                    )

    g.addedge(fromnode="A", tonode=tonode, 
            style = edgestyle,
            # label=['$\beta_{jk}$', '$\beta_{kj}$'],
            #   lw = 20,
            label = [r'$\beta_{BA}$',betalabel],
            labeloffset = edgelabeloffset,
            labelfontsize=30*fontscale,
            loopkwargs = {'lw':edgelw}
            )

    g.draw(debug=False,overfrac = 2,figsize=figsize)


    # g.save("./figs/testABstar.svg")
    return g

ABgraph()
```
